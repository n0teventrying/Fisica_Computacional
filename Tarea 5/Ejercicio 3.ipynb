```python
# a)
#si, es raro que ande importando todo de a poquito a poquito
#pero los que saben dicen que es mejor hacerlo así.
from pylab import plot, show, xlabel, ylabel,title
from math import exp

c = 2
exactitud = 1e-6

x = 1.0 #valor incial
error = 1.0 # 

while error > exactitud:
	x,xprima = 1 - exp(-c*x), x 
	error = abs((x-xprima)/(1-1/(c*exp(-c*x)))) #Expresión para el error en el Newman
print(x)
```

```python
# b)
import numpy as np 

#si bien es desde 0, si lo corremos así, nuestro padrino el python dice 'cuidadito crack'
C = np.arange(0.1,3,0.1)
X = []
for c in C:	
	x = 1.0
	error = 1.0
	while error > exactitud:
		x,xprima = 1 - exp(-c*x), x
		error = abs((x-xprima)/(1-1/(c*exp(-c*x))))
	X.append(x)	
title('X vs C')
xlabel('C')
ylabel('X')
plot(X,C)
show()

#No comento mucho en estos porque no fueron retadores como los anteriores (lit eso solo transcribir lo del Newman en
#algunos casos) y así va a ser para la mayoria que sigue.
#Incluso diria que el 2 también fue como estos, pero por los loops no tanto.
```
