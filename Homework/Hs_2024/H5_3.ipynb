```python
# a)
from pylab import*
from math import exp

c = 2
exactitud = 1e-6

x = 1.0 #valor incial
error = 1.0 # 

while error > exactitud:
	x,xprima = 1 - exp(-c*x), x 
	error = abs((x-xprima)/(1-1/(c*exp(-c*x)))) #Expresión para el error en el Newman
print(x)
```

```python
# b)
import matplotlib.pyplot as plt
import numpy as np 

C = np.arange(0.1,3,0.1)
X = []
for c in C:	
	x = 1.0
	error = 1.0
	while error > exactitud:
		x,xprima = 1 - exp(-c*x), x
		error = abs((x-xprima)/(1-1/(c*exp(-c*x))))
	X.append(x)	
plt.plot(X,C)
plt.show()
```
