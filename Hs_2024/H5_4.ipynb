```python
# b)
from pylab import*

def f(a,b,x,y):
    
    x1 = x
    y1 = y
    
    x = y1*(a+x1**2)
    y = b/(a+x1**2)
    
    return x,y

a = 1
b = 2

x0 = 1.
y0 = 1.

x,y = x0,y0
for n in range(70):
    x,y = f(a,b,x,y)
    print(x,y)
```
