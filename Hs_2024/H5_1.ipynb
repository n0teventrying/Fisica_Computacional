LU DECOMPOSITION.

Me gustaria poner aquí todo el desarrollo de como se llegan a las matrices LU, pero estamos de acuerdo de que aquí en la notebook sería muy tedioso.

Bueno, la idea de usar la descomposición de LU, es que podemos tener una matriz inicial A y un conjunto de vectores $v_{i}$, y para todos los vectores se comparte esa matriz, lo que queremos hacer es como grabar todas esas operaciones y aplicarlas solo al vector v.

Si compactamos todas las operaciones en una secuencia de matrices llegamos a que:
$$
L_{i}L_{i-1}...L_{1}L_{0}Ax = L_{i}L_{i-1}...L_{1}L_{0}v
$$

Pero bueno, si definimos a $L = L_{0}L_{1}...L_{i-1}L_{i}$ y $U = L_{i}L_{i-1}...L_{1}L_{0} A$ entonces $LU = A$, y por lo tanto:
$$ 
LUx = v
$$
Si definimos a $Ux = y$, entonces $Ly = v$
Lo que no mencione y se veria claramente si hubiera puesto a las matrices explicitamente es que L es una matriz triangular inferior y la U una matriz triangular superior.
Por lo que para resolver este sistema, primero debemos hacer una Foward substitution, es decir:

$$l_{00}y_{0} = v0  $$

Por el momento no vamos a mencionar las matrices explicitamente, solo es el procedimiento por encima, si desean ver todos los desarrollos están en el Newman y en mis notas de github.
